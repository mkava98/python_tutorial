```python
data = np.random.normal(0,30,1000)
bin_size = np.arange(-100,100,10) #start #stop #step
plt.hist(data, bins = bin_size, alpha = 0.9 )
plt.title('random gausian data')
plt.xlabel('bin size')
plt.ylabel('count')
plt.show()
```

##pie

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```
