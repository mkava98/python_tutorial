## Pandas DataFrames

```python
home1 = {'room' :2,
          'area' : 250,
          'warehouse' : 0}

home2 = {'room' :1,
          'area' : 80,
          'warehouse' : 1}

home3 = {'room' :3,
          'area' : 550,
          'warehouse' : 1}
```

```python
df = pd.DataFrame([home1, home2, home3], index = ['home1', 'home2', 'home3'])
df
```

```python
df.loc['home2', 'area']
```

```python
df.loc[:, 'room']
```

```python
df.drop('home2')
```

```python
df
```

```python
df['area'].count()
```
